```python
import numpy as np
import pandas as pd
wine = pd.read_csv('/Users/wlee/Desktop/Georgia Tech/Spring 2018/CS 4400 Intro to Databases/Project/wine-reviews/winemag-data-130k-v2.csv')
wine.head()
```

```python
wine = wine[['country', 'description', 'title', 'designation', 'points', 'price', 'province', 'region_1', 'region_2', 'variety', 'winery']]
wine.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sort_country = wine.groupby(['country']).count()['title'].sort_values(ascending = False)
sort_country.head(10)
sns.barplot(sort_country.head(10).values, sort_country.head(10).index)
```

```python
sort_province = wine.groupby(['province']).count()['title'].sort_values(ascending = False)
sort_province.head(10)
sns.barplot(sort_province.head(10).values, sort_province.head(10).index)
```

```python
sort_variety = wine.groupby(['variety']).count()['title'].sort_values(ascending = False)
sort_variety.head(10)
sns.barplot(sort_variety.head(10).values, sort_variety.head(10).index)
```

```python
sort_winery = wine.groupby(['winery']).count()['title'].sort_values(ascending = False)
sort_winery.head(10)
sns.barplot(sort_winery.head(10).values, sort_winery.head(10).index)
```

```python
df_rg = wine.copy()
df_rg['country'] = wine.country.astype('category')
df_rg['variety'] = wine.variety.astype('category')
df_rg['winery'] = wine.winery.astype('category')
df_rg.head()
```

```python
top_variety = pd.DataFrame(sort_variety.head(10).index)
df_variety = df_rg.merge(top_variety, on='variety', how="inner")
df_variety.head()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn import preprocessing
import seaborn as sns
%matplotlib inline

top_country = pd.DataFrame(sort_country.head(10).index)
df_country = df_rg.merge(top_country, on='country', how="inner")

predictor = 'country'

model = smf.ols('points ~ {}'.format(predictor), data = df_country).fit()

params = model.params.drop(index="Intercept")

sns.barplot(x = params.values, y = params.index)
model.summary()
```

```python
top_province = pd.DataFrame(sort_province.head(10).index)
df_province = df_rg.merge(top_province, on='province', how="inner")

predictor = 'province'

model = smf.ols('points ~ {}'.format(predictor), data = df_province).fit()

params = model.params.drop(index="Intercept")

sns.barplot(x = params.values, y = params.index)
model.summary()
```

```python
top_variety = pd.DataFrame(sort_variety.head(10).index)
df_variety = df_rg.merge(top_variety, on='variety', how="inner")

predictor = 'variety'

model = smf.ols('points ~ {}'.format(predictor), data = df_variety).fit()

params = model.params.drop(index="Intercept")

sns.barplot(x = params.values, y = params.index)
```

```python
top_winery = pd.DataFrame(sort_winery.head(10).index)
df_winery = df_rg.merge(top_winery, on='winery', how="inner")

predictor = 'winery'

model = smf.ols('points ~ {}'.format(predictor), data = df_winery).fit()

params = model.params.drop(index="Intercept")

sns.barplot(x = params.values, y = params.index)

model.summary()
```

```python
df_price = wine[['title', 'points', 'price']]
df_price.head()
```

```python
model = smf.ols('points ~ price', data = wine).fit()

from statsmodels.graphics.regressionplots import abline_plot

# scatter-plot data
ax = wine.plot(x='price', y='points', kind='scatter')

# plot regression line
abline_plot(model_results=model, ax = ax)

model.summary()
```

```python
df_scale = df_rg[['title', 'points', 'price']]
df_scale = df_price.dropna(how='any')
df_scale['sc_price'] = preprocessing.scale(df_scale['price'])
df_scale['sc_points'] = preprocessing.scale(df_scale['points'])

model = smf.ols('sc_points ~ sc_price', data = df_scale).fit()

from statsmodels.graphics.regressionplots import abline_plot

# scatter-plot data
ax = df_scale.plot(x='sc_price', y='sc_points', kind='scatter')

# plot regression line
abline_plot(model_results=model, ax = ax)

model.summary()
```

```python
top_variety = pd.DataFrame(sort_variety.head(10).index)
df_variety = df_rg.merge(top_variety, on='variety', how="inner")[['title', 'points', 'price', 'variety']]
df_variety = df_variety.dropna(how='any')
df_variety['sc_price'] = preprocessing.scale(df_variety['price'])

predictor = 'sc_price + variety'

model = smf.ols('points ~ {}'.format(predictor), data = df_variety).fit()

params = model.params.drop(index="Intercept")

sns.barplot(x = params.values, y = params.index)
model.summary()
```
